```python
# Import libraries
import pandas as pd
import numpy as np
pd.options.display.max_columns = None
pd.set_option.chained_assignment = None
```

```python
# Read the data
data = pd.read_csv('glassdoor_jobs.csv')
```

```python
data.head()
```

# Preprocess the Salary Estimate column

```python
data['Salary Estimate'].unique()
```

```python
data['Salary Estimate'].value_counts()
```

```python
data = data[data['Salary Estimate'] != "-1"]
```

```python
data['Salary Estimate'].value_counts()
```

```python
# Replace any words/phrases in the parentheses with space
import re
pattern = '(\(.+\))'
salary = data['Salary Estimate'].apply(lambda x: re.sub(pattern, '', x).strip())
```

```python
salary.value_counts()
```

```python
# Replace the "K" and "$" characters with space
salary = salary.apply(lambda x: x.replace('K', '').replace('$', ''))
```

```python
salary.value_counts()
```

```python
# Create two new columns with boolean values if the position's salary is hourly-based or provided by employer
data['hourly'] = data['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
data['employer_provided'] = data['Salary Estimate'].apply(lambda y: 1 if 'employer provided salary' in y.lower() else 0)
```

```python
data.head()
```

```python
salary.value_counts()
```

```python
salary_range = salary.apply(lambda x: x.lower().replace('per hour', '').replace('employer provided salary:', '').strip())
```

```python
salary_range.value_counts()
```

```python
# Get average salary
data['min_salary'] = salary_range.apply(lambda x: int(x.split('-')[0]))
data['max_salary'] = salary_range.apply(lambda x: int(x.split('-')[1]) if len(x.split('-')) == 2 else int(x.split('-')[0]))
```

```python
data['avg_salary'] = (data['min_salary'] + data['max_salary']) / 2
```

```python
data.head()
```

# Preprocessing the Company Name column

```python
data['Company Name'].unique()
```

```python
# Split each value at new line character to get the company name
data['company_name'] = data['Company Name'].apply(lambda x: x.split('\n')[0])
```

# Preprocessing the Location column

```python
data['Location'].unique()
```

```python
data['Location'] = data['Location'].replace(np.nan, 'Unknown')
```

```python
# Create a job_state column with state value from the Location
data['job_state'] = data['Location'].apply(lambda x: x.split(',')[1] if len(x.split(',')) == 2 else x.split(',')[0])
```

```python
data.job_state.unique()
```

# Preprocessing the Age column

```python
data['age'] = data.Founded.apply(lambda x: x if x < 1 else 2022 - x)
```

# Preprocessing the Job Description column

```python
data['Job Description'][0]
```

```python
# Replace any character which is not a word, a numeric character and some special characters with space
pattern = '[^a-zA-Z0-9\s.,-/:]'
data['new_description'] = data['Job Description'].apply(lambda x: re.sub(pattern, '', x))
```

```python
data.head()
```

```python
# A list of most used technologies and skills in data science
tool_lists = ['python', 'r studio', 'r-studio', 'spark', 'aws', 'excel', 'matlab', 'tableau', 'powerbi', 'sql']
```

```python
for tool in tool_lists:
    data['tool_{}'.format(tool)] = data['Job Description'].apply(lambda x: 1 if tool in x.lower() else 0)
```

```python
data[data['tool_excel'] == 1].head()
```

# Replace unknown and NaN values 

```python
data.Size.value_counts()
```

```python
data['Size'] = data['Size'].replace('-1', 'Unknown')
```

```python
data.Size.value_counts()
```

```python
pattern = '(\d+)'
data['company_size'] = data['Size'].str.extract(pattern, expand = False)
data['company_size'] = data['company_size'].fillna('Unknown')
data['company_size'] = data['company_size'].apply(lambda x: 'Unknown' if x == 'Unknown' else str(x)+'+')
```

```python
data['company_size'].value_counts(dropna = False)
```

```python
data.company_size.isnull().sum()
```

```python
data['Type of ownership'].value_counts()
```

```python
data['Type of ownership'] = data['Type of ownership'].replace('-1', 'Unknown')
```

```python
data['Type of ownership'].value_counts()
```

```python
data['Type of ownership'] = data['Type of ownership'].replace(np.nan, 'Unknown')
```

```python
data['Type of ownership'].value_counts()
```

```python
data['Industry'].value_counts()
```

```python
data['Industry'] = data['Industry'].replace('-1', 'Unknown')
```

```python
data['Industry'].isnull().sum()
```

```python
data['Industry'].value_counts()
```

```python
data['Revenue'].value_counts()
```

```python
usd_pattern = '\(\w+\)'
data['revenue'] = data['Revenue'].apply(lambda x: re.sub(usd_pattern, '', x))
data['revenue'] = data['revenue'].apply(lambda x: str(x))
```

```python
data['revenue'] = data['revenue'].replace('-1', 'Unknown / Non-Applicable')
```

```python
data['revenue'].value_counts()
```

```python
data['revenue'].isnull().sum()
```

```python
data_out = data.drop(['Salary Estimate', 'Job Description', 'Company Name', 'Location', 'Size', 'Sector', 'Revenue'], axis = 1)
```

```python
data_out.head()
```

```python
# Rename the columns for better exploration
data_out.columns = data_out.columns.str.lower().str.strip().str.replace(' ', '_')
```

```python
# Convert the dataframe to csv file
data_out.to_csv('final_data.csv', index = None)
```
